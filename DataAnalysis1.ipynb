21Ô∏è. Identify missing values in the dataset.

22Ô∏è. Fill missing Age with mean value.

```python
#Fill NaN values with mean of Age Column
df['Age']=(
    df['Age']
    .fillna(df['Age'].mean())
)
df[df['Age'].isna()] #Check is there any NaN value left in Age column
```

23Ô∏è. Fill missing City with the most frequent value.

```python
#Fill the NaN values in the city column with most repeated(mode) 'City' name.
df['City']=(
    df['City']
    .fillna(df['City'].mode()[0])
)
df[df['City'].isna()]
```

26. Standardize Gender column (male, MALE, Male ‚Üí Male,(fmale, fALE, female ‚Üí Female)

```python
df['Gender']=(
    df['Gender']
    .str.strip() #Remove extra spaces before and after the string
    .str.capitalize() #Makes 1st letter capital, and remaining small.
    .replace({'M':'Male','F':'Female'}) #Replaces the strings based on starting letters
)
df['Gender'].head() #To check
```

üü° LEVEL 2 ‚Äì Data Filtering (Easy)

1Ô∏è‚É£ Get all employees whose salary is greater than 40,000.

```python
# df['Salary']>40000  #Using this you might get some error. It is because your column{Salary} has some non numerical value/values.
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce') #Converts any non-numerical to NaN values
df[df['Salary']>40000] #Prints all the rows and columns of employees whose salary is greater than specified one.
```

```python
df['Salary'].isna().sum() #Prints number of NaN values present in the specified column.
```

2Ô∏è‚É£ Filter employees who work in the ‚ÄúSales‚Äù department.

```python
df[df['Department']=='Sales']
#Print the dataframe of values of Deparment column From dataframe {df} in the column 'Department' where values equal to "Sales".
```

3Ô∏è‚É£ Select male employees who belong to ‚ÄúDelhi.‚Äù

```python
df[(df['Gender']=='Male') & (df['City']=='Delhi')]
 #Show all employees who are Male AND live in Delhi.
```

4Ô∏è‚É£ Get employees with Age > 25 and Salary < 40,000.


```python
df[(df['Age']>25) & (df['Salary']<50000)]
#From dataframe df,
#take the rows where the value in column 'Age' is greater than 25
#AND
#from dataframe df, the value in column 'Salary' is less than 50,000.
```

5Ô∏è‚É£ Get rows where City is missing (NaN).

```python
df[df['City'].isna()] # .isna() is used to find the NaN or Null values in the specified column
#Here we have no NaN value in the dataframe
```

```python
df['City'].isna().sum() #Optional-- To re-check if we have any NaN value present in the column
```

6Ô∏è‚É£ Show all employees who joined after 2022.

```python
# df[df['Join_Date'] > '2022-12-31'] #You get an error
#because the column 'Join_Date' has different formats like
#["12-05-2023", "2022/11/10", "2023.04.01", "2021", "abc", NaN values]
df['Join_Date']=pd.to_datetime(df['Join_Date'],errors='coerce', dayfirst=True)
#Convert Join_Date to proper datetime
df[df['Join_Date'] > '2022-12-31'] #Filter employees who joined after 2022
```
